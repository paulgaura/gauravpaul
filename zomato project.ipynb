# Filling Null values accordingly

```python
df.address=df.address.fillna("unknown")
```

```python
df.zipcode=df.zipcode.fillna('unknown')
```

```python
df['cuisines']=df['cuisines'].fillna('no descrption')
```

```python
df.timings=df.timings.fillna('11 AM to 11 PM')
```

```python
df.opentable_support=df.opentable_support.fillna(0.0)
```

# TOP 10 names 

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(6, 15))



# Plot the total crashes
sns.set_color_codes("pastel")
sns.barplot(x=df.name.value_counts().head(10), y=df.name.value_counts().head(10).index,label="Total", color="b")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 500), ylabel="Top 10 names",
       xlabel="no.outlet")
sns.despine(left=True,bottom=True)
```

# Average rating of top 15 outlet

```python
avrage_rat=df.groupby("name")["aggregate_rating"].mean().sort_values(ascending=False)
```

```python
rat=avrage_rat[df['name'].value_counts().index].head(15).round(1)
```

```python
rat
```

```python
sns.barplot(x=rat,y=rat.index)
rat.reset_index()
```

# Restaurants with high number of vote

# TOP 10 CITY 
NUMBER OF RESTARUANTS


```python
top10 = df.city.value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top10.values,y=top10.index)

plt.title('Distribution of Restaurants Across Top 10 Cities')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.grid(axis='y')
plt.show()
```
